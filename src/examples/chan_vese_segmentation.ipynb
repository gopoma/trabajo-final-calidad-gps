```python
import os
import sys
sys.path.append("../") 

import cv2
import matplotlib.pyplot as plt

from enhance import ChanVeseSegmentation
```

```python
def work(filepath):
    info = ChanVeseSegmentation(filepath)

    image = cv2.imread(filepath)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    plt.imshow(image), plt.xticks([]), plt.yticks([])
    plt.contour(info[0], [0], colors="r", linewidth=2)
    plt.draw(), plt.show(block=False), plt.pause(0.5)
```

```python
filenames = ["bear.jpg", "car_2.bmp", "car_3.bmp", "car_4.bmp", "fighter.jpg", "gourd.bmp", "hist_0.jpg"]
for filename in filenames:
    work(os.path.join("..", "assets", filename))
```
